# **Q8:** Ek list `numbers = [12, 45, 7, 23, 56, 90, 11]` mein se sabse bada (Maximum) number find karein loop ke zariye.

````python
numbers = [12, 45, 7, 23, 56, 90, 11]

max_num = numbers[0]

for num in numbers:
    if num > max_num:
        max_num = num

print ("sabse bada number hai: ", max_num)
````

# **Q18:** Ek manual counting karein ke ek string (e.g., 'Learning Python is fun') mein kitne **Vowels** (a, e, i, o, u) hain.

````python
text = input("string dalooo:")
vowels = "aeiouAEIOU"
count = 0

for char in text :
    if char in vowels:
        count = count + 1

        print ("total vowels hain :" , count)
````

# **Q20:** Ek 'Number Guessing Game' banayein jahan program ek number (1-20) fix kare aur user ko 3 chances de guess karne ke liye. `while loop` aur `break` ka use karein.

````python
secreat_num = 7 
chances = 3

print("mainne 1 se 20 tak ek number sochahai. guess karo!")

while chances > 0:
    guess = int(input("apna guess daalo:"))

    if guess == secreat_num:
        print("sahi jawab! tum jeet gaye ")
        break
    elif guess < secreat_num:
        print("bohot chota hai")
    else:
        print("bohhot bada hai!")

        chances = chances - 1
        print ("bache hue chances:" , chances)

if chances == 0:
    print("game over! sahi number tha:" ,secreat_num)

````
